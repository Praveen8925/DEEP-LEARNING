```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,MinMaxScaler,OneHotEncoder
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sn
```

```python
d1=pd.read_csv('//content/Churn_Modelling.csv')
d1.sample(10)
```

```python
d1.drop(['RowNumber','CustomerId','Surname','Geography'],axis=1,inplace=True)
```

```python
d1.dtypes
```

```python
for col in d1:
  if d1[col].dtype=='object':
    print(f'{col}:{d1[col].unique()}')
```

```python
#d1=pd.get_dummies(data,columns=['Geography'])
```

```python
lb_processing=LabelEncoder()
for col in d1:
  if d1[col].dtype=='object':
    d1[col]=lb_processing.fit_transform(d1[col])
  if d1[col].dtype=='bool':
    d1[col]=lb_processing.fit_transform(d1[col])
```

```python
scale=MinMaxScaler()
col=['Age','CreditScore','Tenure','Balance','NumOfProducts','EstimatedSalary']
d1[col]=scale.fit_transform(d1[col])


#d1=scale.fit_transform(d1[['Age','CreditScore','Tenure','Balance','NumOfProducts','EstimatedSalary']])
```

```python
d1.dtypes
```

```python
d1.sample(5)
```

```python
x=d1.drop('Exited',axis=1)
y=d1['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train[:10]
```

```python
x_test[:10]
```

```python
model=keras.Sequential([
    keras.layers.Dense(12,input_shape=(9,),activation='relu'),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp=model.predict(x_test)
yp[:10]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:20]
```

```python
y_test[:20]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
