```python
#SPLIT THE DATA INTO TRAING AND TESTING
(x_train,y_train),(x_test,y_test)=data
```

```python
#SCLAE THE X_TRAIN AND X_TEST BETWEEN 0 TO 1
x_train=x_train/255
x_test=x_test/255
```

```python
#ONEHOTENCODING FOR Y_TRAIN AND Y_TEST

y_train_catogrical=keras.utils.to_categorical(
    y_train,10).astype('float32')
y_test_catogrical=keras.utils.to_categorical(
    y_test,10).astype('float32')
```

```python
#MODEL TRAING
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(3000,activation='relu'),
    keras.layers.Dense(1000,activation='relu'),
    keras.layers.Dense(500,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train_catogrical,epochs=30)
```

```python
model.evaluate(x_test,y_test_catogrical)
```
